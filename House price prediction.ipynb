```python
# Use an available style
plt.style.use('seaborn-v0_8-bright')

# Histogram: Distribution of house prices
plt.figure(figsize=(10, 6))
plt.hist(df['price'], bins=50, color='skyblue', edgecolor='black')
plt.title('Histogram of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Pie Chart: Proportion of properties by condition
condition_counts = df['condition'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(condition_counts, labels=condition_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Property Conditions')
plt.axis('equal')
plt.show()
```

```python
# Line Chart: Average price trend over time
df['date'] = pd.to_datetime(df['date'])
price_trend = df.groupby(df['date'].dt.to_period('M'))['price'].mean()
price_trend.index = price_trend.index.to_timestamp()

plt.figure(figsize=(12, 6))
plt.plot(price_trend.index, price_trend.values, marker='o', linestyle='-')
plt.title('Average House Price Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
